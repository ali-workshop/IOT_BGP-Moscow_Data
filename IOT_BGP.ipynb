```python
import networkx as nx
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a NetworkX graph
G = nx.Graph()

# Add nodes and edges to represent your network in a smaller grid pattern
num_rows = 3
num_cols = 3
spacing = 2.0  # Adjust this value to control node spacing

for row in range(num_rows):
    for col in range(num_cols):
        node_name = f"router_{row}_{col}"
        G.add_node(node_name)
        if row > 0:
            G.add_edge(node_name, f"router_{row-1}_{col}")
        if col > 0:
            G.add_edge(node_name, f"router_{row}_{col-1}")

# Define anomalous nodes
anomalous_nodes = ["router_0_0", "router_2_2"]

# Ensure that all anomalous nodes are added to the graph
for node in anomalous_nodes:
    if not G.has_node(node):
        G.add_node(node)

# Create a layout for the graph (Kamada-Kawai layout)
pos = nx.kamada_kawai_layout(G, dim=2)

# Adjust node positions for increased spacing
for node in pos:
    x, y = pos[node]
    pos[node] = (x * spacing, y * spacing)

# Assign a fixed Z-coordinate to create a 3D effect
for node in pos:
    pos[node] = (pos[node][0], pos[node][1], 0)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw all nodes in 3D
x, y, z = zip(*[pos[node] for node in G.nodes()])
ax.scatter(x, y, z, c='lightblue', s=500, label='Normal Node')

# Draw edges in 3D
for edge in G.edges():
    x_edge, y_edge, z_edge = zip(*[pos[node] for node in edge])
    ax.plot(x_edge, y_edge, z_edge, color='lightblue', linewidth=1, alpha=0.7)

# Highlight anomalous nodes in 3D
x_anomalous, y_anomalous, z_anomalous = zip(*[pos[node] for node in anomalous_nodes])
ax.scatter(x_anomalous, y_anomalous, z_anomalous, c='red', s=800, label='Anomalous Node')

# Set 3D axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set plot title and legend
plt.title("3D Network Anomalies Detected")
ax.legend()

# Show the 3D plot
plt.show()
```
